```python
#取引先名でグルーピング
companyGroupedData = masterData.groupby("取引先名")
```

```python
#各値のsumを出力
companyGroupedData.sum()
```

```python
#assign(列名 = 値)メソッドで列を追加、値を代入した
kr12cross = companyGroupedData.sum().assign(PS30 = 0).assign(一面3 = 0)
kr12cross
```

```python
#PS30を基準にカテゴリ分けを行った
kr12cross["PS30"] = kr12cross["プラチナスカウト送信数"].apply(lambda x: 0 if x < 30 else 1)

#1面3名を基準にカテゴリ分けを行った
kr12cross["一面3"] = kr12cross["一次面接数"].apply(lambda x: 0 if x < 3 else 1)

kr12cross
```

```python
#クロス計算を実行
objectKr12 = pd.crosstab(kr12cross["PS30"], kr12cross["一面3"])
objectKr12
```
